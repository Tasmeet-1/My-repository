```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
df = pd.read_csv('/content/russia_losses_equipment.csv')
df['date'] = pd.to_datetime(df['date'])
print("Dataset Overview")
print(df.info())
print("\nDataset Columns:", df.columns)
print("\nMissing Values:", df.isnull().sum())
df.select_dtypes(include=['number']).hist(figsize=(10,10))
plt.title("Histograms of Equipment Losses")
plt.show()
sns.lineplot(x="date", y="tank", data=df)
plt.title("Feature Correlation Heatmap")
plt.show()
plt.figure(figsize=(12,6))
plt.plot(df['date'], df['tank'], label="Tank Losses")
plt.plot(df['date'], df['aircraft'], label="Aircraft Losses")
plt.plot(df['date'], df['APC'], label="APC Losses")
plt.xlabel("Date")
plt.ylabel("Count")
plt.title("Equipment Losses Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
df['Loss_rate_tank'] = df['tank'] / df['day'].sum()
df['Loss_rate_aircraft'] = df['aircraft'] / df['day'].sum()
df['Loss_rate_APC'] = df['APC'] / df['day'].sum()
df['Loss_rate_helicopter'] = df['helicopter'] / df['day'].sum()
df['Loss_rate_other'] = df['cruise missiles'] / df['day'].sum()
if "greatest losses direction" in df.columns:
    label_encoder = LabelEncoder()
    df["greatest_losses_direction_encoded"] = label_encoder.fit_transform(df["greatest losses direction"])
    df.drop(columns=["greatest losses direction"], inplace=True)
X = df.drop(columns=["date", "day"])
y = df["tank"]
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
rf = RandomForestRegressor(n_estimators=100, max_depth=None, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("\nFirst 5 Predictions:", y_pred[:5])
```
