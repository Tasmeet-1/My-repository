```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
df = pd.read_csv('/content/appstore_games.csv')
#Emcoding
label_encoder = LabelEncoder()
df['Genres'] = label_encoder.fit_transform(df['Genres'])
df['Primary Genre'] = label_encoder.fit_transform(df['Primary Genre'])
#Normalization of Numerical Columns: -
standard_scaler = StandardScaler()
df['Average User Rating'] = standard_scaler.fit_transform(df['Average User Rating'].values.reshape(-1, 1))
df['User Rating Count'] = standard_scaler.fit_transform(df['User Rating Count'].values.reshape(-1, 1))
#removing outliers: -
def remove_outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
  return df
numerical_cols = ['Average User Rating', 'User Rating Count', 'Price']
for column in numerical_cols:
  df = remove_outliers_iqr(df, column)
  print(df.shape)
print(df.head())
print(df.info())
print(df.describe())
print(df['In-app Purchases'].isnull().sum())
df.dropna(inplace=True)
sns.boxplot(df['Average User Rating'])
plt.show()
#Post Outlier removal
sns.histplot(df['Average User Rating'], bins=30, kde=True)
plt.show()
print(df.dtypes)
numerical_cols = ['Average User Rating', 'User Rating Count', 'Price']
df[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols])
#Feature_Engineering : -
df['Total_Downloads'] = df['Average User Rating'] * df['User Rating Count']
print("Total amount of Downloads:")
print(df['Total_Downloads'])
sns.pairplot(df)
plt.show()
#univariate analysis: -
sns.histplot(df['Average User Rating'], kde=True)
plt.show()
sns.histplot(df['User Rating Count'], kde=True)
plt.show()
sns.histplot(df['Price'], kde=True)
plt.show()
#bivariate analysis: -
sns.scatterplot(x='Average User Rating', y='User Rating Count', data=df)
plt.show()
sns.scatterplot(x='Average User Rating', y='Price', data=df)
plt.show()
sns.scatterplot(x='User Rating Count', y='Price', data=df)
plt.show()
#Multivariate Analysis: -
sns.pairplot(df[['Average User Rating','User Rating Count','Price']])
plt.show()
```
